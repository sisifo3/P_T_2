```python
#En esta seccion vamos a crae los primeros centrides de manera aleatoria 
def Centroides(A,C):
  t = []
  for i in range(C):
    r = [0] * 2
    a = len(A)
    b = len(A[0])
    r1 =  random.randint(1,a)
    r2 = random.randint(1,b)
    r[0] = r1
    r[1] = r2
    t.append(r)
  return  t

C = 9
Cs = Centroides(A,C)    
print(Cs) 
```
