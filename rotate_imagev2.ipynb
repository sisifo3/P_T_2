```python
height, width = img.shape
print(height)
print(width)

a = math.floor(width/2)
print(a)

b = math.floor(height/10)
print(b)

c = math.floor((width/2)/10)
print(c)

croppedImage = img[b: (height-b), (a+c):(width-c)] 
```
